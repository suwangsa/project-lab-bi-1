```python
import glob
path = 'datasets/'
list_folder = ['balita','bayi','hamil','kekerasan','faskes','umum']
all_tables = {}

for folder in list_folder:
    all_zip = []
    all_zip = glob.glob(path + folder + '/' + '*.zip')
    all_tables[folder] = all_zip
```

```python
from zipfile import ZipFile

for folder in list_folder:
    for file_name in all_tables[folder]:
        with ZipFile(file_name, 'r') as zip:
            # printing all the contents of the zip file
            csv_file = zip.namelist()[0]
            pathway = (path + folder)

            # extracting all the files
            zip.extract(csv_file,pathway)
```

```python
import glob
import re
import os
pathway = 'datasets/umum/data.jabarprov.go.id/'

def dinkes(file_path):
    all_files = glob.glob(file_path + '*.csv')
    list_files = [pd.read_csv(filename,index_col=None, header=0).copy() for filename in all_files]

    col_to_delete = ['kode_provinsi','nama_provinsi','id','kode_kabupaten_kota','satuan']
    done_list = []
    
    file_name = [re.search(r"(?<=od\_)(\S+)(?=\.csv)",file).group() for file in all_files]
    save_path = re.search(r"(\S+)(?=data.jabarprov.go.id)",pathway).group()

    for num,files in enumerate(list_files):
        for col in files.columns:
            if files[col].isna().any()==False:
                if re.match(r"persen", col):
                    files[col].astype('float64')
                elif re.match(r"jumlah", col):
                    files[col].astype('int64')
            
            if col in col_to_delete:
                files.drop(columns=col,inplace=True)
            
            
        files.to_csv(os.path.join(save_path, file_name[num] + '.csv'))

```
